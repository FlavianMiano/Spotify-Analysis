# Spotify Analysis

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from wordcloud import WordCloud
```

```python
# loading the dataset

df = pd.read_csv('SpotifyFeatures.csv')
df.head(5)
```

```python
artists = pd.read_csv('artists.csv')
artists.head(5)
```

```python
tracks = pd.read_csv('tracks.csv')
tracks.head(5)
```

```python
# checking for null values

tracks.isnull().sum()
```

```python
# checking the info

tracks.info()
```

```python
sort_tracks = tracks.sort_values('popularity', ascending= True).head(10)
sort_tracks
```

```python
# describe

tracks.describe().transpose()
```

```python
most_pop = tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_pop[:10]
```

```python
#setting release date as index
tracks.set_index('release_date', inplace = True)

#converting to date and time format
tracks.index = pd.to_datetime(tracks.index)

tracks.head(5)
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x : round(x/1000))
tracks = tracks.drop(columns='duration_ms')
tracks.head(5)
```

```python
# correlation

corr = tracks.corr()

plt.figure(figsize=(10,7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr ,annot=True,cmap="inferno",fmt=".2f", mask=mask);
plt.title('Correlation between the columns')
plt.show()
```

```python
# creating a sample dataset

sample_df = tracks.sample(int(0.004*len(tracks)))
sample_df.head(5)
```

```python
sample_df.shape
```

```python
# correlation between loudness and energy

plt.figure(figsize=(10, 6))

sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = 'Loudness vs energy correlation')
```

```python
# correlation between loudness and energy

plt.figure(figsize=(10, 6))

sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'b').set(title = 'popularity vs acousticness correlation')
```

```python
tracks['year'] = tracks.index.get_level_values('release_date')
tracks.year = pd.to_datetime(tracks.year).dt.year
tracks.head(5)
```

```python
# year disribution

plt.figure(figsize=(20,8))

tracks['year'].value_counts().plot(kind = 'bar', xlabel = 'Year',
                                   ylabel = 'Count', title = 'Distribution of music throughout the years')
```

```python
sns.displot(tracks['year'], discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of songs per year')
```

```python
sns.set_style(style = 'whitegrid')
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = tracks.year, y = tracks.duration, ax = ax).set(title = 'Year vs Duration')
```

```python
plt.figure(figsize=(10,5))

tracks['artists'].value_counts()[:20].plot(kind = 'bar', xlabel = ' Artist Name',
                                        ylabel = 'count', title = 'Number of songs per Artist')
```

```python
tracks['name'].dtype
```

```python
# Concatenate all text data into one string
text = (str(tracks['name']))

# Generate the wordcloud
wordcloud = WordCloud(width=800, height=800,
                      background_color='white',
                      min_font_size=10).generate(text)

plt.figure(figsize=(6, 10), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.title('Common words used in music titles')
plt.tight_layout(pad=0)

plt.show()
```

```python
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df)
plt.title('Duration of music per Genre')
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10, 5))
pop = df.sort_values('popularity', ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = pop).set(title = 'Top 5 genres')
```

```python
pop_artist = artists.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='name', x = 'popularity', data = pop_artist).set(title = 'Artist vs Popularity')
```
